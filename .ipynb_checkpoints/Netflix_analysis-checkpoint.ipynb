```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt

# Read in the CSV as a DataFrame
df_csv=pd.read_csv('datasets/netflix_data.csv')

# Print the first five rows of the DataFrame
df_csv.head(5)
```

##  Filtering for movies!
<p>Scanning the column, we see there are also TV shows in the dataset! so we will filter them to study only movies</p>

```python
# Subset the DataFrame for type "Movie"
df_csv_movie = df_csv[df_csv['type']=='Movie']

# Select only the columns of interest
df_movie= df_csv_movie.loc[:,['title', 'country', 'genre', 'release_year', 'duration']]

# Print the first five rows of the new DataFrame
df_movie.head(5)
```

## Digging for movies shorter than 60 minutes
<p>Let's filter our DataFrame for movies with a <code>duration</code> under 60 minutes and look at the genres. </p>

```python
# Filter for durations shorter than 60 minutes
df_short_movie=df_movie[df_movie['duration']<60]

# Print the first 20 rows of short_movies
df_short_movie.head(20)
```

##  Marking non-feature films
<p>It looks as many of the films that are under 60 minutes fall into genres such as "Children", "Stand-Up", and "Documentaries". </p>
<p>WE will plot them, but mark them with a different color generating a list of colors based on the contents of the <code>genre</code> column. </p>

* children: red

* documentaries: blue

* stand-up: green

## Plotting with color!


```python
# Define an empty list
lista=[]

# Iterate over rows of df_movie
for label, row in df_movie.iterrows():
    if row['genre']=='Children':
        lista.append('red')
    elif row['genre']=='Documentaries':
        lista.append('blue')
    elif row['genre']=='Stand-Up':
        lista.append('green')
    else:
        lista.append('gray')


# Set the figure style and initalize a new figure
plt.style.use('fivethirtyeight')
fig=plt.figure(figsize=(12,8))

# Create a scatter plot of duration versus release_year
plt.scatter(df_movie['release_year'],df_movie['duration'], c=lista,label='All genres, except Children, Documentaries & Stand-Up')

# Create a title and axis labels
plt.title('Duration & Release year')
plt.xlabel('Release year')
plt.ylabel('Duration (minutes)')
plt.legend()

# Show the plot
plt.show()
```

##  Are movies getting shorter?

#### No, only genres as children's movies (red), documentaries (blue) and stand-up (green) have a short duration, 
#### Most of the movies with other genres (gray)  are longer than an hour

## What were the most popular genres?

#### We count the number of movies for each genre

```python
# Get values form genre column
genres = df_movie['genre'].values

# Count the number of movies for each genre
count_genres={}
for genre in genres:
    if genre in count_genres.keys():
        count_genres[genre] += 1
    else:
        count_genres[genre] = 1
print(count_genres)
```

### We will sort in descending order 

```python
# using sorted function with the tuple: "diccionario.items()" 
tuple_sorted= sorted(count_genres.items(), key=lambda x: x[1], reverse=True)

# Crea un nuevo diccionario con los valores ordenados en orden descendente.
dictionary_sorted = {clave: valor for clave, valor in tuple_sorted}
dictionary_sorted
```

## Ojiva Plot

<p>We will apply ojiva technique to identify in wich genres are the 80% of the movies</p>

```python
# Get genres from dictionary in a list
genres=list(dictionary_sorted.keys())

# Get values from dictionary in a list
absolute_frequency = list(dictionary_sorted.values())
```

## Getting calculation to plot the "Ojiva"

* Relative frequency
* Absolute cumulative frequency
* Relative cumulative frequency

```python
# Calculating relative_frequency using the absolute_frequency
relative_frequency = []

for i in range(len(absolute_frequency)):
    af = absolute_frequency[i]/sum(absolute_frequency)
    relative_frequency.append(af)
    


# Calculating absolute_cumulative_frequency using the absolute_frequency
absolute_cumulative_frequency = []

summary = 0
for no in range(len(absolute_frequency)):
    acf=absolute_frequency[no]+summary
    absolute_cumulative_frequency.append(acf)
    summary=acf
    
    
    
# Calculating relative_cumulative_frequency using the relative_frequency
relative_cumulative_frequency = []

summary = 0
for no in range(len(absolute_frequency)):
    rcf=relative_frequency[no]+summary
    relative_cumulative_frequency.append(rcf)
    summary=rcf
```

## Build dataframe

```python
# Creatre a dictionary with the calculations into the lists
dictionary = {'genres':genres,
              'absolute_frequency':absolute_frequency,
              'relative_frequency':relative_frequency,
              'absolute_cumulative_frequency':absolute_cumulative_frequency,
              'relative_cumulative_frequency':relative_cumulative_frequency}

# Creating dataframe
df_ojiva = pd.DataFrame(dictionary)
df_ojiva
```

## Plotting

```python
fig, ax = plt.subplots()

ax.plot(df_ojiva.index, df_ojiva['relative_cumulative_frequency'])
ax.plot(4, 0.824, marker='o', markersize=10)

plt.text(5, 0.824, '82.4%', bbox = {'facecolor': 'oldlace', 'alpha': 0.9, 'boxstyle': "larrow,pad=0.3", 'ec': 'red'})

ax.set_ylabel('Relative cumulative frequency')
ax.set_title('"Ojiva" Plot ')

plt.show()
```

### We have 19 genres but only 5 genres represents the 82% of total, the popular genres are:
* Dramas
* Comedies
* Doumentaries
* Action
* Children

#### In the plot we can see how the line slope is high and after 82% begins to decrease
